```python
#read data from file and handle null values
df = pd.read_csv('orders.csv', na_values = ['Not Available','unknown'])
```

```python
df['Ship Mode'].unique()
```

```python
#rename column names...make them lower case and replace with underscore
#df.rename (columns = {'Order Id': 'order_id', 'City':'city'}) not a good practice
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.columns
```

```python
df.head(4)
```

```python
#derive new columns discount, sale price and profit
df['discount'] = df['list_price'] * df['discount_percent']/100
```

```python
df.head(5)
```

```python
df['sale_price'] = df['list_price'] - df['discount']
```

```python
df.head(5)
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head(4)
```

```python
df.dtypes
```

```python
#convert order_date from string datatype to datetime
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
#drop cost_price, list_proce and discount_percent columns
df.drop(columns=['list_price', 'cost_price', 'discount_percent']) # However this does not make changes in the original df
```
